# 日産コンテスト 画像認識サンプルコード
TensorFlow2系とKerasを用いて画像分類モデルの構築と訓練、学習結果の保存を行うサンプルコードです。

参考： [TensorFlowチュートリアル Image classification](https://www.tensorflow.org/tutorials/images/classification?hl=ja)

## ライブラリのインポート

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Flatten, Dropout, MaxPooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import PIL
import cv2
import time
import numpy as np
import matplotlib.pyplot as plt
```

## データの理解と準備

```python
# 画像のあるディレクトリの指定
train_dir = '../../train'
train_parking_dir = os.path.join(train_dir, 'parking')  # 学習用の parking 画像のディレクトリ
train_street_dir = os.path.join(train_dir, 'street')  # 学習用の street 画像のディレクトリ
```

```python
# 画像枚数の確認
total_parking=len(os.listdir(train_parking_dir))
print('Total training parking images: {}'.format(total_parking))
total_street=len(os.listdir(train_street_dir))
print('Total training street images: {}'.format(total_street))
total_train=total_parking+total_street
print('Total training images: {}'.format(total_train))
```

```python
# パラメータの設定
batch_size = 128
epochs = 5
IMG_WIDTH = 640  # 元画像は1920
IMG_HEIGHT = 360  # 元画像は1080
```

```python
# 学習データのジェネレータの作成
train_image_generator = ImageDataGenerator(rescale=1./255)

train_data_gen = train_image_generator.flow_from_directory(batch_size=batch_size,
                                                           directory=train_dir,
                                                           classes=["parking", "street"],
                                                           shuffle=True,
                                                           target_size=(IMG_HEIGHT, IMG_WIDTH),
                                                           class_mode='binary')
```

## データの可視化

```python
sample_training_images, _ = next(train_data_gen)
```

```python
# この関数は、1行5列のグリッド形式で画像をプロットし、画像は各列に配置されます。
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 5, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
plotImages(sample_training_images[:5])
```

## モデルの作成

```python
# モデルの構築

def create_model():
    model = Sequential([
        Conv2D(8, 3, padding='same', activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH ,3)),
        MaxPooling2D(),
        Conv2D(16, 3, padding='same', activation='relu'),
        MaxPooling2D(),
        Conv2D(32, 3, padding='same', activation='relu'),
        MaxPooling2D(),
        Flatten(),
        Dense(64, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    
    return model
```

```python
# モデルのコンパイル
model=create_model()
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# モデルの概要の表示
model.summary()
```

## 学習

```python
# モデルの学習
history = model.fit(
    train_data_gen,
    steps_per_epoch=total_train // batch_size,
    epochs=epochs,
#     validation_data=val_data_gen,
#     validation_steps=total_val // batch_size
)
```

## 学習結果の可視化

```python
epochs_range = range(epochs)
acc = history.history['accuracy']
# val_acc = history.history['val_accuracy']

loss = history.history['loss']
# val_loss = history.history['val_loss']

plt.figure(figsize=(1, 2))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
# plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
# plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```
